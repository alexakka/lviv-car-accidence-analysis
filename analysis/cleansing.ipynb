У значеннях стовпця можна побачити невідповідності в регістрах літер та лишні пробіли. Маємо такі дублікати як 'перекидання тз' та 'наїзд на опори'. Також є орфографічні помилки 'накїзд' замість 'наїзд', 'пішоходний' замість 'пішохідний'

```python
import re

PUNCT_TO_DELETE = str.maketrans('', '', '()')

REPLACEMENTS = {
    "накїзд": "наїзд",
    "піш.переходу": "пішохідного переходу",
    "пішоходний": "пішохідний",
    "моб. зв'язку": "мобільного зв'язку",
    "транспортний засіб": "тз",
}

def normalize_accident(s: str) -> str:
    if not isinstance(s, str):
        return s

    s = s.lower().translate(PUNCT_TO_DELETE)

    for old, new in REPLACEMENTS.items():
        s = s.replace(old, new)

    s = ' '.join(s.split())
    return s
```

```python
accidents_df['accidentTypeNorm'] = accidents_df['accidentType'].apply(normalize_accident)

unq_accident_types = accidents_df["accidentTypeNorm"].unique()

print(f"Кількість унікальних значень: {len(unq_accident_types)}\n")

for accident_type in sorted(unq_accident_types):
    print(accident_type)
```

Розбіжності в правописі та регістрах вирішено

У стовпці присутнє велика кількість дублювання інформації. Так, до прикладу, значення 'наїзд на пішохода', 'зіткнення', 'наїзд на велосипедиста', 'наїзд на транспортний засіб' мають чималу кількість коментарів. Для таких випадків можна додати новий стовпець 'accidentTypeComment' в якому міститимуться коментарі, щодо типу аварій.

```python
BASE_PATTERNS = (
    r'зіткнення',
    r'наїзд на пішохода',
    r'наїзд на велосипедиста',
    r'наїзд на транспортний засіб, що стоїть',
    r'наїзд на опори'
)

base_alt   = '|'.join(BASE_PATTERNS)
rx_comment = re.compile(rf'^({base_alt})(?:\s*-\s*|\s+)(.+)$')
```

```python
def split_type_comment(s: str):
    s = s.strip()

    # 1) є дефіс " - "
    if ' - ' in s:
        base, comment = s.split(' - ', 1)
        return base.strip(), comment.strip()

    # 2) немає дефіса ‒ пробуємо regex
    m = rx_comment.match(s)
    if m:
        return m.group(1).strip(), m.group(2).strip()

    # 3) зовсім без уточнення
    return s, 'N/A'
```

```python
accidents_df[['accidentTypeBase', 'accidentTypeComment']] = (
    accidents_df['accidentTypeNorm']
      .apply(lambda x: pd.Series(split_type_comment(x)))
)
```

```python
unq_accident_types = accidents_df["accidentTypeBase"].unique()

print(f"Кількість унікальних значень: {len(unq_accident_types)}\n")

for accident_type in sorted(unq_accident_types):
    print(accident_type)
```
